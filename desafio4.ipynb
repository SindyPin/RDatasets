- Largura da sépala;
- Comprimento da pétala;
- Largura da pétala;
- Espécie.

# Lembrando dos diferentes tipos de correlações...

Em relação ao tipo de correlação...

- CORRELAÇÃO POSITIVA: Conforme uma variável aumenta, a outra variável também aumenta;

- CORRELAÇÃO NEGATIVA: Conforme uma variável aumenta, a outra variável diminui;

- CORRELAÇÃO NULA: Sem correlação.

Em relação à intensidade da correlação...

- FORTE: Quanto menor for a dispersão os pontos, maior será a correlação entre os dados;

- FRACA: Quanto maior for a dispersão os pontos, menor será o grau entre os dados.
